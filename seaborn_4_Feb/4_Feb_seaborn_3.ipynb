# seaborn parameter part - 3 

```python
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df= sn.load_dataset("tips")
df.head()
```

```python
# lagend parameter do not have to be given. it takes by default.
sn.lineplot(x="size",y="total_bill", hue="sex",data=df)
```

# style ---------

```python
# style parameter is used to show different lines.
sn.lineplot(x="size",y="total_bill",
            hue="sex",style="sex",data=df)
```

# "palette" we can change colour

```python
# whenever we give any balue in the palette . it will show a error"valueerror"
#then we have show possible value are:Accent,Accent_r .....etc.
# all type of value we can give there is show
sn.lineplot(x="size",y="total_bill", 
            hue="sex",palette="Accent_r",data=df)
# can give a lot of value to a palette
```

```python
#"deshes" its use without dashes lines will drow
sn.lineplot(x="size",y="total_bill", 
            hue="sex",dashes=False,data=df)
```

# Markers 

```python
sn.lineplot(x="size",y="total_bill", 
            hue="sex",dashes=False,style="sex",
            markers=["o","<"],data=df,palette="hot")
```

```python
#legend` must be 'brief', 'full', or False
# we can set the fontsize of the lagend itself.
sn.lineplot(x="size",y="total_bill",legend="full", 
            hue="sex",data=df)
plt.xlabel("Size", fontsize=15)
plt.ylabel("Total Bill",fontsize=15)
plt.title("Line Plot", fontsize=15)
```

```python

#style must be one of white, dark, whitegrid, darkgrid, ticks
sn.set(style="dark")
sn.lineplot(x="size",y="total_bill",legend="full", 
            hue="sex",data=df)
plt.xlabel("Size", fontsize=15)
plt.ylabel("Total Bill",fontsize=15)
plt.title("Line Plot", fontsize=15)
```
