# how to draw  seaborn histogram/seaborn Distplot  

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df=sn.load_dataset("tips")
df
```

```python
sn.distplot(df["size"])
```

```python
sn.distplot(df["tip"])
```

```python
sn.distplot(df["total_bill"])
```

```python
# How many parameters we can give 
'''
sn.distplot(
    a,# "a" means jo data huma dana ha for ex:df["total_bill"] 
    bins=None,
    hist=True,
    kde=True,
    rug=False,
    fit=None,
    hist_kws=None,
    kde_kws=None,
    rug_kws=None,
    fit_kws=None,
    color=None,
    vertical=False,
    norm_hist=False,
    axlabel=None,
    label=None,
    ax=None,
'''
```

```python
sn.distplot(df["total_bill"],bins=50)
```

```python
sn.distplot(df["total_bill"],hist=False)
# if we need a only line we can use hist=false, by default=True
```

```python
sn.distplot(df["total_bill"],kde=False)
# if we need a only histogram we can use
```

```python
sn.distplot(df["total_bill"],rug=True)
```

```python
from scipy.stats import norm
sn.distplot(df["total_bill"],fit=norm)
# there are two line plots foud here
# but we need only one we will use "kde=False"
```

```python
# we can say that we should normal line plot
sn.distplot(df["total_bill"],fit=norm,kde=False)
# we can take more value in place of norm
```

# color 

```python
sn.distplot(df["total_bill"],color="r")
```

```python
# if we want to visualize the histogram as a horizontal shape we will use verical=True
sn.distplot(df["total_bill"],vertical=True)
```

```python
# if we need to change xlabel name  so we can use 
sn.distplot(df["total_bill"],axlabel="Total Bill")
```

```python
sn.distplot(df["total_bill"],label="Total Bill")
plt.legend()
```
